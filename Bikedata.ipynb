```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
april_2023_trips = Path("Resources/CSV Files/JC-202303-citibike-tripdata.csv")
may_2023_trips = Path("Resources/CSV Files/JC-202304-citibike-tripdata.csv")
june_2023_trips = Path("Resources/CSV Files/JC-202305-citibike-tripdata.csv")
july_2023_trips = Path("Resources/CSV Files/JC-202306-citibike-tripdata.csv")
august_2023_trips = Path("Resources/CSV Files/JC-202307-citibike-tripdata.csv")

#Read the CSV & create databases
apr_trips = pd.read_csv(april_2023_trips)
may_trips = pd.read_csv(may_2023_trips)
jun_trips = pd.read_csv(june_2023_trips)
jul_trips = pd.read_csv(july_2023_trips)
aug_trips = pd.read_csv(august_2023_trips)

#Showing the first few rows to check if the data is loaded
may_trips.head()
```

```python
#Combining all the data to a single database
result_df = pd.concat([apr_trips, may_trips,
                       jun_trips, jul_trips, aug_trips], axis=0, ignore_index=True)

result_df.head()
```

```python
#Exporting to an Excel file
result_df.to_excel('Resources/Source Data.xlsx', index=False)
```
